## Editing Data

```python
#creating new (2) cols for the data that is a percentage of the population that is either unger 18 and over 65

Mar_age["under_18"] = (
        Mar_age["Male: Under 5 years"] + Mar_age["Male: 5 to 9 years"] + Mar_age["Male: 10 to 14 years"] + Mar_age["Male: 15 to 17 years"] +
        Mar_age["Female: Under 5 years"] + Mar_age["Female: 5 to 9 years"] + Mar_age["Female: 10 to 14 years"] + Mar_age["Female: 15 to 17 years"]
    )
Mar_age["over_65"] = ( Mar_age["Male: 65 and 66 years"] + Mar_age["Male: 67 to 69 years"] +
        Mar_age["Male: 70 to 74 years"] + Mar_age["Male: 75 to 79 years"] + Mar_age["Male: 80 to 84 years"] + Mar_age["Male: 85 years and over"] +
        Mar_age["Female: 65 and 66 years"] + Mar_age["Female: 67 to 69 years"] +
        Mar_age["Female: 70 to 74 years"] + Mar_age["Female: 75 to 79 years"] + Mar_age["Female: 80 to 84 years"] + Mar_age["Female: 85 years and over"]
    )
Mar_age["under_18_per"] = Mar_age["under_18"]/Mar_age["Total"]
Mar_age["over_65_per"] = Mar_age["over_65"]/Mar_age['Total'] 

Elk_age["under_18"] = (
        Elk_age["Male: Under 5 years"] + Elk_age["Male: 5 to 9 years"] + Elk_age["Male: 10 to 14 years"] + Elk_age["Male: 15 to 17 years"] +
        Elk_age["Female: Under 5 years"] + Elk_age["Female: 5 to 9 years"] + Elk_age["Female: 10 to 14 years"] + Elk_age["Female: 15 to 17 years"]
    )
Elk_age["over_65"] = ( Elk_age["Male: 65 and 66 years"] + Elk_age["Male: 67 to 69 years"] +
        Elk_age["Male: 70 to 74 years"] + Elk_age["Male: 75 to 79 years"] + Elk_age["Male: 80 to 84 years"] + Elk_age["Male: 85 years and over"] +
        Elk_age["Female: 65 and 66 years"] + Elk_age["Female: 67 to 69 years"] +
        Elk_age["Female: 70 to 74 years"] + Elk_age["Female: 75 to 79 years"] + Elk_age["Female: 80 to 84 years"] + Elk_age["Female: 85 years and over"]
    )
Elk_age["under_18_per"] = Elk_age["under_18"]/Elk_age["Total"]
Elk_age["over_65_per"] = Elk_age["over_65"]/Elk_age['Total'] 

StJoe_age["under_18"] = (
        StJoe_age["Male: Under 5 years"] + StJoe_age["Male: 5 to 9 years"] + StJoe_age["Male: 10 to 14 years"] + StJoe_age["Male: 15 to 17 years"] +
        StJoe_age["Female: Under 5 years"] + StJoe_age["Female: 5 to 9 years"] + StJoe_age["Female: 10 to 14 years"] + StJoe_age["Female: 15 to 17 years"]
    )
StJoe_age["over_65"] = ( StJoe_age["Male: 65 and 66 years"] + StJoe_age["Male: 67 to 69 years"] +
        StJoe_age["Male: 70 to 74 years"] + StJoe_age["Male: 75 to 79 years"] + StJoe_age["Male: 80 to 84 years"] + StJoe_age["Male: 85 years and over"] +
        StJoe_age["Female: 65 and 66 years"] + StJoe_age["Female: 67 to 69 years"] +
        StJoe_age["Female: 70 to 74 years"] + StJoe_age["Female: 75 to 79 years"] + StJoe_age["Female: 80 to 84 years"] + StJoe_age["Female: 85 years and over"]
    )
StJoe_age["under_18_per"] = StJoe_age["under_18"]/StJoe_age["Total"]
StJoe_age["over_65_per"] = StJoe_age["over_65"]/StJoe_age['Total'] 
```

```python
Mar_inc = Mar_inc.rename(columns = {"Median Income: " : "Median_Income"})
Elk_inc = Elk_inc.rename(columns = {"Median Income:" : "Median_Income"})
StJoe_inc = StJoe_inc.rename(columns = {"Median Income: "  : "Median_Income"})
```

```python
Mar_pov["NAME"] = Mar_pov["NAME"].astype('object') 
Elk_pov["NAME"] = Elk_pov["NAME"].astype('object') 
StJoe_pov["NAME"] = StJoe_pov["NAME"].astype('object')

Mar_age["NAME"] = Mar_age["NAME"].astype('object') 
Elk_age["NAME"] = Elk_age["NAME"].astype('object') 
StJoe_age["NAME"] = StJoe_age["NAME"].astype('object')

Mar_sex["NAME"] = Mar_sex["NAME"].astype('object') 
Elk_sex["NAME"] = Elk_sex["NAME"].astype('object') 
StJoe_sex["NAME"] = StJoe_sex["NAME"].astype('object')

Mar_inc["NAME"] = Mar_inc["NAME"].astype('object') 
Elk_inc["NAME"] = Elk_inc["NAME"].astype('object') 
StJoe_inc["NAME"] = StJoe_inc["NAME"].astype('object')

#Making all NAME cols be the same data type
```

```python
county_fips = ["099", "141", "039"] #Marshall, St. Joe, and Elkart

in_counties_filt = in_counties[in_counties['COUNTYFP'].isin(county_fips)]
```

```python
Elkhart = ["039"]
filtered_counties1 = in_counties_filt[in_counties_filt['COUNTYFP'].isin(Elkhart)]

Marshall = ["099"]
filtered_counties3 = in_counties_filt[in_counties_filt['COUNTYFP'].isin(Marshall)]

St_Joe = ["141"]
filtered_counties2 = in_counties_filt[in_counties_filt['COUNTYFP'].isin(St_Joe)]

#creates three subsets of df where we are only based on one county
```

## Building Map Data

```python
Elkh = pd.merge(filtered_counties1, Elk_pov[['NAME', 'Poverty_Level']], on='NAME', how='left')
Elkha = pd.merge(Elkh, Elk_inc[["NAME", "Median_Income"]], on = "NAME", how ='left')
Elkhart = pd.merge(Elkha, Elk_age[["NAME", "under_18_per", "over_65_per"]], on = "NAME", how ='left')
Elkhart["Median_Income"] = Elkha["Median_Income"].str.replace("$","").str.replace(",","").astype(float) #needed as it base has the dollar and comma which messes with folium
```

```python
StJose = pd.merge(filtered_counties2, StJoe_pov[['NAME', 'Poverty_Level']], on='NAME', how='left')
StJosep = pd.merge(StJose, StJoe_inc[["NAME", "Median_Income"]], on = "NAME", how ='left')
StJoseph = pd.merge(StJosep, StJoe_age[["NAME", "under_18_per", "over_65_per"]], on = "NAME", how ='left')
```

```python
Mars = pd.merge(filtered_counties3, Mar_pov[['NAME', 'Poverty_Level']], on='NAME', how='left')
Marsh = pd.merge(Mars, Mar_inc[["NAME", "Median_Income"]], on = "NAME", how ='left')
Marshall = pd.merge(Marsh, Mar_age[["NAME", "under_18_per", "over_65_per"]], on = "NAME", how ='left')
```

```python
countiesthree = pd.concat([Elkhart, StJoseph, Marshall]) 
countiesthree["Poverty_Level"] = countiesthree["Poverty_Level"].str.replace("%","").astype(float) #Remove Percent sign and makes value a float
countiesthree["under_18_per"] = round(countiesthree["under_18_per"]*100,2) #Not in decimal form for ease of reading
countiesthree["over_65_per"] = round(countiesthree["over_65_per"]*100,2) #Not in decimal form for ease of reading
```

## Map

```python
mf = fm.Map(location=[41.5, -86],
                zoom_start = 9)
fm.Choropleth(
    name = "Poverty Rate",
    geo_data=countiesthree,  
    data=countiesthree,      
    columns=['GEOID', 'Poverty_Level'], 
    key_on='feature.properties.GEOID', 
    fill_color='YlOrRd',   
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Census Tract Poverty Rate (%)',
    highlight=True
).add_to(mf)

fm.Choropleth(
    name = "Median Income",
    geo_data=countiesthree,  
    data=countiesthree,      
    columns=['GEOID', 'Median_Income'],
    key_on='feature.properties.GEOID', 
    fill_color='RdYlGn',    
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Census Tract Median Income',
    highlight=True,
).add_to(mf)

fm.Choropleth(
    name = "Under 18 Percent",
    geo_data=countiesthree,  
    data=countiesthree,      
    columns=['GEOID', 'under_18_per'],
    key_on='feature.properties.GEOID', 
    fill_color='BuPu',    
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Census Tract Under 18 Percentage',
    highlight=True,
).add_to(mf)

fm.Choropleth(
    name = "Over 65 Percent",
    geo_data=countiesthree,  
    data=countiesthree,      
    columns=['GEOID', 'over_65_per'],
    key_on='feature.properties.GEOID', 
    fill_color='RdBu',    
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Census Tract Over 65 Percentage',
    highlight=True,
).add_to(mf)

fm.GeoJson(
    data=countiesthree,
    name='Census Tract Info',
    style_function = lambda feature : {
        'weight' : .5, 
        'fillOpacity': 0.0,
        'color' : 'black'
    },
    tooltip=fm.GeoJsonTooltip(
        fields=["NAME", "Poverty_Level","Median_Income","under_18_per", "over_65_per"],  
        aliases=['Census Tract:', 'Poverty Rate:', 'Median Income:',"Percentage of Population Under 18:", "Percentage of Population Over 65:"],
        localize=True,
        sticky=False
    )
).add_to(mf)

#Bus route code came from Cheyenne as I did not know how and was running out of time to complete the map

routes = gpd.read_file("TranspoRoutes.shp").to_crs(epsg=4326)

route_colors = {
    "1 Madison / Mishawaka": "navy",
    "10 Western Avenue": "turquoise",
    "11 Southside Mishawaka": "maroon",
    "12 Rum Village": "midnightblue",
    "12/14 Rum Village / Sample": "thistle",
    "13 Corby / Town & Country": "yellow",
    "14 Sample / Mayflower": "mediumorchid",
    "15A University Park Mall / Mishawaka (via Main Stree)": "saddlebrown",
    "15B University Park Mall / Mishawaka (via Grape Road)": "burlywood",
    "16 Blackthorn Express": "hotpink",
    "17 The Sweep": "olivedrab",
    "3A Portage": "firebrick",
    "3B Portage": "red",
    "4 Lincolnway West / Excel Center / Airport": "darkorange",
    "5 North Michigan / Laurel Woods": "navy",
    "6 South Michigan / Erskine Village": "red",
    "7 Notre Dame / University Park Mall": "forestgreen",
    "7A Notre Dame Midnight Express": "green",
    "8 Miami / Scottsdale": "turquoise",
    "8/6 Miami / Scottsdale / South Michigan / Erskine Vi": "red",
    "9 Northside Mishawaka": "magenta"
}

routes["color_name"] = routes["line_name"].map(route_colors).fillna("gray")

def route_style(feature):
    name = feature["properties"]["line_name"]
    color = route_colors.get(name, "#808080")  #default gray
    return {"color": color, "weight": 3, "opacity": 0.9}

routes_group = fm.FeatureGroup(name="Transportation Routes")
fm.GeoJson(
    routes,
    style_function=route_style,
    tooltip=fm.GeoJsonTooltip(
        fields=["line_name", "full_name", "color_name"],
        aliases=["Line:", "Full Name:", "Line Color:"]
    )
).add_to(routes_group)
mf.add_child(routes_group)

fm.LayerControl(show = False).add_to(mf)

mf

mf.save("Python_Map_Recreation.html")
```
