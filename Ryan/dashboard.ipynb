```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("Elkhart_PovPercent_Transposed.csv")

# Clean column names
df.columns = [col.strip() for col in df.columns]

# Helper to parse percentages
def parse_pct(val):
    if isinstance(val, str):
        return float(val.strip('%'))
    return val

# Identify the columns
total_poverty_col = 'Income in the past 12 months below poverty level:'
df[total_poverty_col] = df[total_poverty_col].apply(parse_pct)

# Sort by poverty rate for better visualization
df_sorted = df.sort_values(by=total_poverty_col, ascending=False)

# Total Poverty Rate by Census Tract
plt.figure(figsize=(12, 8))
sns.barplot(x='NAME', y=total_poverty_col, data=df_sorted, palette='viridis')
plt.xticks(rotation=90)
plt.title('Poverty Rate by Census Tract in Elkhart')
plt.ylabel('Poverty Rate (%)')
plt.xlabel('Census Tract')
plt.tight_layout()
plt.savefig('poverty_rate_by_tract.png')

# Age Group Analysis (Aggregated for all tracts)
# Get male and female age columns
age_cols_male = [col for col in df.columns if 'Male:' in col]
age_cols_female = [col for col in df.columns if 'Female:' in col]

for col in age_cols_male + age_cols_female:
    df[col] = df[col].apply(parse_pct)

# Sum averages across all tracts to see which age group is most affected
age_summary = df[age_cols_male + age_cols_female].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
age_summary.plot(kind='barh', color='skyblue')
plt.title('Average Poverty Rate Contribution by Age and Gender')
plt.xlabel('Average Poverty Rate (%)')
plt.ylabel('Age Group / Gender Category')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('poverty_by_age_gender.png')

print("Charts generated: poverty_rate_by_tract.png and poverty_by_age_gender.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean the data again
df_outbound = pd.read_csv('Outbound.csv')

# Numeric conversion
num_cols = ['Protein LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS', 'Total Pounds']
for col in num_cols:
    df_outbound[col] = pd.to_numeric(df_outbound[col], errors='coerce').fillna(0)

# Date and Year/Month extraction
df_outbound['Date'] = pd.to_datetime(df_outbound['Date'], errors='coerce')
df_outbound = df_outbound.dropna(subset=['Date'])
df_outbound['Year'] = df_outbound['Date'].dt.year
df_outbound['Month'] = df_outbound['Date'].dt.month

# Filter out 2026
df_filtered = df_outbound[df_outbound['Year'] < 2026].copy()

# Clean County names
df_filtered['County'] = df_filtered['County'].astype(str).str.strip().str.upper()

# 1. Monthly Trend (2023-2025)
monthly_all = df_filtered.groupby(['Year', 'Month'])['Total Pounds'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_all, x='Month', y='Total Pounds', hue='Year', marker='o', palette='tab10')
plt.title('Monthly Food Distribution Trend (2023 - 2025)', fontsize=14)
plt.xticks(range(1, 13))
plt.ylabel('Total Pounds')
plt.tight_layout()
plt.savefig('trend_23_25.png')

# 2. Category Comparison (2023-2025)
categories = ['Protein LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS']
cat_sums = df_filtered.groupby('Year')[categories].sum().reset_index().melt(id_vars='Year', var_name='Category', value_name='Pounds')
cat_sums['Category'] = cat_sums['Category'].str.replace(' LBS', '')
plt.figure(figsize=(12, 6))
sns.barplot(data=cat_sums, x='Category', y='Pounds', hue='Year', palette='viridis')
plt.title('Comparison of Food Categories (2023 - 2025)', fontsize=14)
plt.ylabel('Total Pounds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('category_23_25.png')

# 3. County Distribution (Top 5)
county_stats = df_filtered.groupby(['County', 'Year'])['Total Pounds'].sum().reset_index()
major_counties = county_stats.groupby('County')['Total Pounds'].sum().sort_values(ascending=False).head(5).index
county_stats_filtered = county_stats[county_stats['County'].isin(major_counties)]
plt.figure(figsize=(10, 6))
sns.barplot(data=county_stats_filtered, x='County', y='Total Pounds', hue='Year', palette='magma')
plt.title('Food Distribution by County (Top 5, 2023-2025)', fontsize=14)
plt.ylabel('Total Pounds')
plt.tight_layout()
plt.savefig('county_23_25.png')

# 4. Top 10 Vendors (2025 Only)
latest_full_year = df_filtered[df_filtered['Year'] == 2025]
top_vendors_25 = latest_full_year.groupby('Vendor')['Total Pounds'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 8))
sns.barplot(data=top_vendors_25, x='Total Pounds', y='Vendor', palette='rocket')
plt.title('Top 10 Food Recipients in 2025 (Full Year)', fontsize=14)
plt.xlabel('Total Pounds')
plt.tight_layout()
plt.savefig('vendors_2025.png')

print("2026 data removed. Graphs regenerated for 2023-2025.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Outbound.csv")

# Reload if necessary (already in memory)
df['Date'] = pd.to_datetime(df['Date'])
df['Total Pounds'] = df['Total Pounds'].fillna(0)

# 1. Total Pounds by Month
df_monthly = df.set_index('Date').resample('M')['Total Pounds'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(df_monthly['Date'], df_monthly['Total Pounds'], marker='o', linestyle='-')
plt.title('Total Pounds Outbound by Month')
plt.xlabel('Month')
plt.ylabel('Total Pounds')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('total_pounds_by_month.png')
plt.show()
plt.close()

# 2. Top 10 Counties by Total Pounds
df_county = df.groupby('County')['Total Pounds'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 6))
plt.bar(df_county['County'], df_county['Total Pounds'], color='skyblue')
plt.title('Top 10 Counties by Total Pounds')
plt.xlabel('County')
plt.ylabel('Total Pounds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_counties.png')
plt.show()
plt.close()

# 3. Total Pounds by Program
df_program = df.groupby('Program')['Total Pounds'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
plt.bar(df_program['Program'], df_program['Program'].index.map(lambda x: df_program.loc[x, 'Total Pounds']), color='salmon')
# Re-mapping to ensure order for bar chart:
plt.clf()
plt.bar(df_program['Program'], df_program['Total Pounds'], color='salmon')
plt.title('Total Pounds by Program (in millions)')
plt.xlabel('Program')
plt.ylabel('Total Pounds in Millions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('total_by_program.png')
plt.show()
plt.close()

# 4. Sum of Food Categories
category_cols = ['Protein LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS', 'Indiv Meal LBS']
categories_sum = df[category_cols].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
categories_sum.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Food Categories (Total Million LBS)')
plt.xlabel('Category')
plt.ylabel('Total Pounds in Millions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('food_categories_distribution.png')
plt.show()
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Outbound.csv')

# Define category columns
category_cols = ['Protein LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS', 'Indiv Meal LBS']

# Calculate sums
categories_sum = df[category_cols].sum()

# Sort categories for better visualization
categories_sum = categories_sum.sort_values(ascending=False)

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(categories_sum, labels=categories_sum.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Food Categories (Total LBS)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

# Save the plot
plt.savefig('food_categories_pie.png')
plt.show()
plt.close()

print(categories_sum)
```

```python
import pandas as pd

# Load the data
df = pd.read_csv('Schools Impact Data.csv')

# Display basic info and first few rows
print(df.info())
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop rows where % Free/Reduced is null for relevant stats
df_clean = df.dropna(subset=['% Free/Reduced'])

# 1. Total number of schools with free/reduced lunch data
total_schools_with_data = len(df_clean)

# 2. Public vs Private distribution
public_private_counts = df_clean['Public'].value_counts()
public_private_pct = df_clean['Public'].value_counts(normalize=True) * 100

# 3. Average Free/Reduced Lunch %
avg_free_reduced = df_clean['% Free/Reduced'].mean()

# 4. Average Free/Reduced by County
avg_by_county = df_clean.groupby('County')['% Free/Reduced'].mean().sort_values(ascending=False)

# 5. Cultivate Support Analysis
cultivate_support_counts = df_clean['Cultivate support '].value_counts()
avg_free_by_support = df_clean.groupby('Cultivate support ')['% Free/Reduced'].mean()

# Plotting
# 1. Public vs Private
plt.figure(figsize=(8, 6))
plt.pie(public_private_counts, labels=public_private_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99'])
plt.title('Distribution of Schools with Free/Reduced Lunch Data (Public vs Private)')
plt.savefig('public_vs_private_pie.png')
plt.show()
plt.close()

# 2. Average % Free/Reduced by County
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_by_county.index, y=avg_by_county.values, palette='viridis')
plt.title('Average % Free/Reduced Lunch by County')
plt.ylabel('Average %')
plt.xlabel('County')
plt.savefig('avg_free_reduced_by_county.png')
plt.show()
plt.close()

# 3. Cultivate Support vs % Free/Reduced
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_free_by_support.index, y=avg_free_by_support.values, palette='magma')
plt.title('Average % Free/Reduced Lunch: Cultivate Support vs No Support')
plt.ylabel('Average %')
plt.xlabel('Cultivate Support (Y/N)')
plt.savefig('cultivate_support_impact.png')
plt.show()
plt.close()

# 4. Distribution of % Free/Reduced Lunch
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['% Free/Reduced'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of % Free/Reduced Lunch Across Schools')
plt.xlabel('% Free/Reduced Lunch')
plt.ylabel('Frequency')
plt.savefig('free_reduced_distribution.png')
plt.show()
plt.close()

# Output statistics
stats = {
    "total_schools_with_data": total_schools_with_data,
    "public_private_counts": public_private_counts.to_dict(),
    "avg_free_reduced": avg_free_reduced,
    "avg_by_county": avg_by_county.to_dict(),
    "cultivate_support_counts": cultivate_support_counts.to_dict()
}
print(stats)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('Schools Impact Data.csv')

# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# 1. Count Private schools with Cultivate support
private_cultivate_support = df[(df['Public'] == 'Private') & (df['Cultivate support'] == 'Y')]
count_private_supported = len(private_cultivate_support)

# 2. Relationship between % Free/Reduced and Cultivate annual meals
# Filter for schools that have data for both
meals_data = df.dropna(subset=['Cultivate annual meals', '% Free/Reduced'])

# Scatter plot: % Free/Reduced vs Cultivate annual meals
plt.figure(figsize=(10, 6))
sns.regplot(x='% Free/Reduced', y='Cultivate annual meals', data=meals_data, scatter_kws={'s':100}, line_kws={'color':'red'})
plt.title('Relationship: % Free/Reduced vs Cultivate Annual Meals')
plt.xlabel('% Free/Reduced')
plt.ylabel('Cultivate Annual Meals')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('meals_vs_poverty_scatter.png')
plt.show()
plt.close()

# Bar chart: Cultivate annual meals by school (for those that have the data)
plt.figure(figsize=(12, 8))
meals_data_sorted = meals_data.sort_values(by='Cultivate annual meals', ascending=False)
sns.barplot(x='Cultivate annual meals', y='School', data=meals_data_sorted, hue='% Free/Reduced', palette='YlOrRd', dodge=False)
plt.title('Cultivate Annual Meals by School (Colored by % Free/Reduced)')
plt.xlabel('Cultivate Annual Meals')
plt.ylabel('School')
plt.legend(title='% Free/Reduced', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('meals_by_school_bar.png')
plt.show()
plt.close()

# Report findings
print(f"Number of Private schools with Cultivate support: {count_private_supported}")
```
