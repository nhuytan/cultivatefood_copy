```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("Elkhart_PovPercent_Transposed.csv")

# Clean column names
df.columns = [col.strip() for col in df.columns]

# Helper to parse percentages
def parse_pct(val):
    if isinstance(val, str):
        return float(val.strip('%'))
    return val

# Identify the columns
total_poverty_col = 'Income in the past 12 months below poverty level:'
df[total_poverty_col] = df[total_poverty_col].apply(parse_pct)

# Sort by poverty rate for better visualization
df_sorted = df.sort_values(by=total_poverty_col, ascending=False)

# Total Poverty Rate by Census Tract
plt.figure(figsize=(12, 8))
sns.barplot(x='NAME', y=total_poverty_col, data=df_sorted, palette='viridis')
plt.xticks(rotation=90)
plt.title('Poverty Rate by Census Tract in Elkhart')
plt.ylabel('Poverty Rate (%)')
plt.xlabel('Census Tract')
plt.tight_layout()
plt.savefig('poverty_rate_by_tract.png')

# Age Group Analysis (Aggregated for all tracts)
# Get male and female age columns
age_cols_male = [col for col in df.columns if 'Male:' in col]
age_cols_female = [col for col in df.columns if 'Female:' in col]

for col in age_cols_male + age_cols_female:
    df[col] = df[col].apply(parse_pct)

# Sum averages across all tracts to see which age group is most affected
age_summary = df[age_cols_male + age_cols_female].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
age_summary.plot(kind='barh', color='skyblue')
plt.title('Average Poverty Rate Contribution by Age and Gender')
plt.xlabel('Average Poverty Rate (%)')
plt.ylabel('Age Group / Gender Category')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('poverty_by_age_gender.png')

print("Charts generated: poverty_rate_by_tract.png and poverty_by_age_gender.png")
```

```python
import pandas as pd
import numpy as np

def clean_data(df, year):
    # Drop rows where Date is null
    df = df.dropna(subset=['Date']).copy()
    
    # Standardize vendor column
    if 'Column3' in df.columns:
        df = df.rename(columns={'Column3': 'Vendor'})
    
    # Date conversion
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date'])
    df['Month'] = df['Date'].dt.month
    df['Year'] = df['Date'].dt.year
    
    # List of numeric columns
    numeric_cols = [
        'Proteins LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 
        'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS', 
        'Indvid Meal LBS', 'Total Pounds'
    ]
    
    for col in numeric_cols:
        if col in df.columns:
            # Remove commas and spaces, then convert to numeric
            df[col] = df[col].astype(str).str.replace(',', '').str.replace(' ', '').replace('nan', '0').replace('', '0')
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
            
    return df

# Clean both dataframes
df_2023_clean = clean_data(pd.read_csv('FoodOutgoing2023_1.csv'), 2023)
df_2025_clean = clean_data(pd.read_csv('FoodOutgoing2025_1.csv'), 2025)

# Combine for comparison
combined_df = pd.concat([df_2023_clean, df_2025_clean], ignore_index=True)

# 1. Total Pounds by Month/Year
monthly_stats = combined_df.groupby(['Year', 'Month'])['Total Pounds'].sum().reset_index()

# 2. Food Category stats (Sums)
category_cols = ['Proteins LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS']
category_stats = combined_df.groupby('Year')[category_cols].sum().reset_index()

# 3. Top Vendors
top_vendors = combined_df.groupby(['Year', 'Vendor'])['Total Pounds'].sum().reset_index()

# 4. County Distribution
county_stats = combined_df.groupby(['Year', 'County'])['Total Pounds'].sum().reset_index()

print("Cleaning and Aggregation Complete.")
print(monthly_stats.head())
print(category_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

# 1. Trend Analysis: Total Pounds by Month
plt.figure(figsize=(10, 6))
# Can filter 2023 to only include months present in 2025 for a fair comparison
sns.lineplot(data=monthly_stats, x='Month', y='Total Pounds', hue='Year', marker='o', palette='tab10')
plt.title('Monthly Food Distribution: 2023 vs 2025 (YTD)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Pounds', fontsize=12)
plt.xticks(range(1, 13))
plt.legend(title='Year')
plt.savefig('food_distribution_trend.png')

# 2. Food Category Breakdown Comparison
# Reshape category_stats for plotting
category_melted = category_stats.melt(id_vars='Year', value_vars=category_cols, var_name='Category', value_name='LBS')
category_melted['Category'] = category_melted['Category'].str.replace(' LBS', '')

plt.figure(figsize=(12, 6))
sns.barplot(data=category_melted, x='Category', y='LBS', hue='Year', palette='muted')
plt.title('Food Distribution by Category (Total Weight)', fontsize=14)
plt.ylabel('Total Pounds', fontsize=12)
plt.xlabel('Food Category', fontsize=12)
plt.xticks(rotation=45)
plt.savefig('food_categories_comparison.png')

# 3. County Distribution (Summing across both years or showing per year)
county_top = combined_df.groupby(['County', 'Year'])['Total Pounds'].sum().reset_index()
# Sort counties by total pounds for better visualization
county_order = combined_df.groupby('County')['Total Pounds'].sum().sort_values(ascending=False).index

plt.figure(figsize=(10, 6))
sns.barplot(data=county_top, x='County', y='Total Pounds', hue='Year', order=county_order, palette='tab10')
plt.title('Total Food Distribution by County', fontsize=14)
plt.ylabel('Total Pounds', fontsize=12)
plt.xlabel('County', fontsize=12)
plt.savefig('county_distribution.png')

# 4. Top 10 Vendors in 2025
top_10_vendors_2025 = df_2025_clean.groupby('Vendor')['Total Pounds'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(data=top_10_vendors_2025, x='Total Pounds', y='Vendor', palette='tab10')
plt.title('Top 10 Food Recipients (Vendors) in 2025', fontsize=14)
plt.xlabel('Total Pounds', fontsize=12)
plt.ylabel('Vendor Name', fontsize=12)
plt.tight_layout()
plt.savefig('top_vendors_2025.png')

print("Visualizations generated.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load datasets
df_2023 = pd.read_csv('FoodOutgoing2023_1.csv')
df_2025 = pd.read_csv('FoodOutgoing2025_1.csv')

def clean_food_data(df, year):
    # Standardize column names
    if 'Column3' in df.columns:
        df = df.rename(columns={'Column3': 'Vendor'})
    
    # Drop rows without a date
    df = df.dropna(subset=['Date']).copy()
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date'])
    
    # Extract Month and Year
    df['Month'] = df['Date'].dt.month
    df['Year'] = year # Use provided year to avoid 2024 dates if any
    
    # Columns to convert to numeric
    numeric_cols = [
        'Proteins LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 
        'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS', 
        'Indvid Meal LBS', 'Total Pounds'
    ]
    
    for col in numeric_cols:
        if col in df.columns:
            # Clean string numbers: remove commas, spaces, etc.
            df[col] = df[col].astype(str).str.replace(r'[^\d.]', '', regex=True)
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
            
    return df

# Clean both
df23 = clean_food_data(df_2023, 2023)
df25 = clean_food_data(df_2025, 2025)

# Combine for cross-year comparison
combined = pd.concat([df23, df25], ignore_index=True)

# Define categories
categories = ['Proteins LBS', 'Starch LBS', 'Veg LBS', 'Fruit LBS', 'Baked Goods LBS', 'Dairy LBS', 'Grocery LBS']

# 1. Monthly Trend Plot
monthly = combined.groupby(['Year', 'Month'])['Total Pounds'].sum().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly, x='Month', y='Total Pounds', hue='Year', marker='o', palette='tab10')
plt.title('Food Outgoing: Monthly Weight Trend (2023 vs 2025)')
plt.ylabel('Total Pounds')
plt.xticks(range(1, 13))
plt.tight_layout()
plt.savefig('food_monthly_trend.png')

# 2. Food Category Comparison
cat_sums = combined.groupby('Year')[categories].sum().reset_index()
cat_melted = cat_sums.melt(id_vars='Year', var_name='Category', value_name='Pounds')
cat_melted['Category'] = cat_melted['Category'].str.replace(' LBS', '')

plt.figure(figsize=(10, 5))
sns.barplot(data=cat_melted, x='Category', y='Pounds', hue='Year', palette='tab10')
plt.title('Comparison of Food Categories (Total Weight)')
plt.ylabel('Pounds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('food_category_comparison.png')

# 3. Top 10 Vendors in 2025
top_vendors = df25.groupby('Vendor')['Total Pounds'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_vendors, x='Total Pounds', y='Vendor', palette='rocket')
plt.title('Top 10 Food Recipients (Vendors) - 2025')
plt.xlabel('Total Pounds')
plt.tight_layout()
plt.savefig('food_top_vendors_2025.png')
```
